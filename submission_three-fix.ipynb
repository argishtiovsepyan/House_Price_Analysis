### Data Cleaning

```python
df_train.columns = df_train.columns.str.lower().str.replace(' ','_')
```

```python
df_train.head()
```

```python
df_test.columns = df_test.columns.str.lower().str.replace(' ', '_')
```

```python
df_test.head()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(df_train.corr()[['saleprice']].sort_values(by = 'saleprice', ascending = False),
            annot = True,
            vmin = -1,
            vmax = 1,
            cmap = 'coolwarm');
```

```python
df_train = df_train.fillna(df_train.mean())
```

```python
X = df_train[['overall_qual',
              'gr_liv_area', 
              'garage_area', 
              'garage_cars', 
              'total_bsmt_sf', 
              '1st_flr_sf', 
              'year_built', 
              'year_remod/add', 
              'full_bath',
              'totrms_abvgrd',
              'fireplaces',
              'bsmtfin_sf_1',
              'open_porch_sf',
              'wood_deck_sf',
              'lot_area'
             ]]
y = df_train['saleprice']
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)
```

```python
X.isna().sum()
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, random_state = 50)
```

```python
print(X_train.shape, y_train.shape)
```

```python
print(X_test.shape, y_test.shape)
```

```python
# y_train = np.array(y_train)
# y_train
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=2023)
np.mean(cross_val_score(lr, X, y, cv=kf))
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=2023)
np.mean(cross_val_score(lr, X_poly, y, cv=kf))
```
